# query_condition

```python
import json
import lolviz as lz

f = open('config/rule_label.json',encoding='utf-8')
user_dic = json.load(f)
f.close()

allabel = []
for key in user_dic.keys():
    group = user_dic[key]
    allabel.append(group['default'])
    if len(group.keys())>3:
        allabel.extend(list(group.keys())[3:])
        
allabel = list(set(allabel))
allabel.remove('')
print(allabel)

f = open('config/hybrid_rule_label.json',encoding='utf-8')
user_dic2 = json.load(f)
f.close()
allabel.extend(list(user_dic2.keys()))
print(allabel)
```

```python
queryjson = {}
for i,label in enumerate(allabel):
    if i%2 == 0:
        #queryjson[label] = {'Timelen':True,'Period':False,'Now':True}
        queryjson[label] = {'constraint':{'Timelen':True,'Period':False,'Now':True},'category':'地区'}
    else:
        #queryjson[label] = {'Timelen':True,'Period':False,'Now':True}
        queryjson[label] = {'constraint':{'Timelen':True,'Period':False,'Now':True},'category':'行业'}
    
print(queryjson)
```

```python
f = open('config/query_condition.json','w',encoding='utf-8')
json.dump(queryjson,f,ensure_ascii=False)
f.close()
f = open('config/query_condition.json',encoding='utf-8')
user_dic3 = json.load(f)
f.close()
print(type(user_dic3))
print(user_dic3)
```
